# Import the necessary libraries

```python
import pandas as pd
import yfinance as yf
```

```python
# Load the Excel file
file_path = 'C:/Users/phitz/OneDrive/Trading Journal.xlsx'
trades_df = pd.read_excel(file_path, sheet_name='Trades')

# Ensure the 'Date' column is of datetime type
trades_df['Date'] = pd.to_datetime(trades_df['Datetime'])

# Display the first few rows of the dataframe to understand its structure
trades_df.head()
```

```python
# Get the minimum date from the trades
min_date = trades_df['Date'].min()

# Extract the unique symbols from the trades
symbols = trades_df['Symbol'].unique().tolist()

# Add benchmark symbol (e.g., 'SPY' for S&P 500)
benchmark_symbol = 'SPY'
symbols.append(benchmark_symbol)

data = yf.download(symbols, start=min_date, end=pd.Timestamp.today(), progress=False)

# Extract the closing prices
closing_prices = data['Close']

# Display the first few rows of the closing prices
closing_prices
```
